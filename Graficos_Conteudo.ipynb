#### **Gráfico de Linha Simples**

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.arange(0,100,0.1)

y = 2*x + 5
z = 2*x**2 + 5*x -2

# Gráfico 1
plt.plot(x,y)
plt.show()

#Gráfico 2
plt.plot(x,y)
plt.grid()
plt.title('Título - Gráfico 2')

plt.show()

#Gráfico 3
plt.plot(x,y, label='y')
plt.plot(x,z, label='z')

plt.grid()
plt.title('Título - Gráfico 3')
plt.legend()

plt.show()
````

##### **Múltiplos Gráficos**

````python
# -- Ordenamento na horizontal

fig, axs = plt.subplots(1,3)

axs[0].plot(x,y)
axs[0].set_title('Gráfico 1')

axs[1].plot(x,y)
axs[1].grid()
axs[1].set_title('Gráfico 2')

axs[2].plot(x,y, label='y')
axs[2].plot(x,z, label='z')
axs[2].grid()
axs[2].legend()
axs[2].set_title('Gráfico 3')

plt.show()
````

````python
# -- Definindo o tamanho das figuras

fig, axs = plt.subplots(1, 3, figsize=(20, 5))

axs[0].plot(x,y)
axs[0].set_title('Gráfico 1')

axs[1].plot(x,y)
axs[1].grid()
axs[1].set_title('Gráfico 2')

axs[2].plot(x,y, label='y')
axs[2].plot(x,z, label='z')
axs[2].grid()
axs[2].legend()
axs[2].set_title('Gráfico 3')

plt.show()
````

````python
w = 3*x**2 + 6*x - 3

fig, axs = plt.subplots(2,2, figsize=(20, 10))

axs[0,0].plot(x,y)
axs[0,0].set_title('Gráfico 1')

axs[0,1].plot(x,y)
axs[0,1].grid()
axs[0,1].set_title('Gráfico 2')

axs[1,0].plot(x,y, label='y')
axs[1,0].plot(x,z, label='z')
axs[1,0].grid()
axs[1,0].legend()
axs[1,0].set_title('Gráfico 3')

axs[1,1].plot(x,z,color= 'red',linestyle = 'dashed', label='z')
axs[1,1].plot(x,w,color= 'green', label='w')
axs[1,1].grid()
axs[1,1].legend()
axs[1,1].set_title('Gráfico 4')

plt.show()
````

````python
#-- Ordenamento na vertical
fig, axs = plt.subplots(2)

axs[0].plot(x,y)
axs[0].set_title('Gráfico 1')
axs[0].set_xlabel('Eixo X')
axs[0].set_ylabel('Eixo Y')

axs[1].plot(x,y)
axs[1].grid()
axs[1].set_title('Gráfico 2')
axs[1].set_xlabel('Eixo X')
axs[1].set_ylabel('Eixo Y')

plt.tight_layout() #Permite inserir todas as características no gráfico sem exceder seu espaço
plt.show()
````

````python
#-- Comando plt.tight_layout()
fig, axs = plt.subplots(2)

axs[0].plot(x,y)
axs[0].set_title('Gráfico 1')
axs[0].set_xlabel('Eixo X')
axs[0].set_ylabel('Eixo Y')

axs[1].plot(x,y)
axs[1].grid()
axs[1].set_title('Gráfico 2')
axs[1].set_xlabel('Eixo X')
axs[1].set_ylabel('Eixo Y')

# plt.tight_layout() #--> Veja que sem o uso deste comando as definições de título do grafico 2 se sobrepoem
# a nomeação e os valores do eixo x do Grafico 1.
plt.show()
````

#### **Gráfico de Dispersão**

````python
x1 = [10, 20, 50, 70, 85, 97]
y1 = [12.45, 13.52, 14.26, 15.87, 16.23, 17.12]

plt.scatter(x1,y1, marker='v', color ='orangered')

plt.grid()
plt.title('Gráfico 4')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
````

````python
#-- Definindo o tamanho dos marcadores (s = n)

x1 = [10, 20, 50, 70, 85, 97]
y1 = [12.45, 13.52, 14.26, 15.87, 16.23, 17.12]

x3 = [10, 20, 30, 40, 50]
y3 = [2, 3, 2, 3, 4]

x2 = [10, 30, 50, 70, 90]
y2 = [6, 8, 7, 8, 7]

plt.scatter(x1,y1, marker='v', color ='darkblue', s= 100)
plt.scatter(x2,y2, marker='>', color ='limegreen', s = 50)
plt.scatter(x3,y3, marker='o', color ='magenta', s = 25)

plt.grid()
plt.title('Gráfico 4')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

plt.savefig('/content/graficos/grafico_4.pdf') #Salvando a figura em pdf
plt.show()
````
