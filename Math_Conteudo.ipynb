###**Biblioteca Math**
___

De uma maneira geral, as biblitoecas são ferramentas que podemos utilizar para complementar as funções que não estão inclusas no python tradicionalmente. Em conteúdos posteriores, veremos como construir funções e as bibliotecas nada mais são de um conjunto de funções de um determinado assunto que nos auxiliam em algumas atividades.

Assim, a biblioteca math é uma das bibliotecas destinadas para funções matemáticas. E, para utiliza-la em seu programa é necessário **importar** a biblioteca para o código através do seguinte comando:

```
import math
```

Normalmente é recomendado relizar a importação da biblioteca e o uso de suas funções em células separadas para não interferir na construção do código. E, caso seja necessário a importação de mais de uma biblioteca, já teremos uma célula destinada apenas para isso.


As bibliotecas geralmente apresentam um "documento" com todas as funções que são compostas apresentando todas as formas corretas de uso e exemplos de suas aplicações. A documentação da biblioteca *math*, por exemplo pode ser  acessada pelo link

> [https://docs.python.org/pt-br/3/library/math.html](https://docs.python.org/pt-br/3/library/math.html)


E, para acessar as funções da biblioteca, é necessário o uso do nome da biblioteca seguida da função que deseja ser acessada. Como por exemplo:
```
# Desta forma, estou atribuindo a variável x o valor da seno de pi
x = math.sin(math.pi)

```





````python
#Célula destinada apenas para importação das bibliotecas
import math
````

````python

x = math.sin(math.pi)
print(x)
````

````python
y = math.floor(x)
print(y)
````

````python
z = math.ceil(x)
print(z)
````

````python
w = math.trunc(x)
print(w)
````

#### **Grau e Radiano**

Além disso, vale ressaltar a transformação de graus para radianos. As funções da biblioteca _Math_ utilizam essencialmente as funções trigonométricas em **radianos**. Assim, caso o problema tenha a entrada dos dados em graus, é possível transformar os valores em radianos e vice e versa com as seguintes funções.


```
#Converte o ângulo x de radianos para graus.
math.degrees(x)   

#Converte o ângulo x de graus para radianos.
math.radians(x)
```

````python
x = 35
x1 = math.radians(x)

print(x1)
````

````python
y = 140
y1 = math.radians(y)

print(y1)
````

````python
z1 = math.degrees(0.6108652381980153)
print(z1)
````

````python
w = 150
w1 = math.sin(math.radians(w))

print(w1)
````
