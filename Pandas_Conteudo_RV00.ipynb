### **Gráficos**


##### **Gráfico de Barras**

````python
#Barras

plt.bar(sepala_M5['class'],sepala_M5['sepal length'])
plt.show()
````

````python
#Histograma

plt.hist(sepala_M5['class'], bins = 5, color='purple')
plt.grid(linewidth=0.5, color = 'lightgray', axis = 'y')
plt.show()
````

##### **Gráfico de Dispersão**

````python
#Dispersão

versicolor = sepala_M5[sepala_M5['class'] =='Iris-versicolor']
setosa = sepala_M5[sepala_M5['class'] =='Iris-setosa']
virginica = sepala_M5[sepala_M5['class'] =='Iris-virginica']

plt.scatter(versicolor['sepal length'],versicolor['sepal width'], label='Versicolor')
plt.scatter(setosa['sepal length'],setosa['sepal width'], label= 'Setosa')
plt.scatter(virginica['sepal length'],virginica['sepal width'], label = 'Virginica')

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
````

````python
plt.scatter(versicolor['sepal length'],versicolor['petal length'], label='Versicolor', color ='deeppink')
plt.scatter(setosa['sepal length'],setosa['petal length'], label= 'Setosa', color='purple')
plt.scatter(virginica['sepal length'],virginica['petal length'], label = 'Virginica', color='deepskyblue')
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.show()
````

````python
plt.scatter(versicolor['sepal length'],versicolor['petal width'], label='Versicolor', color ='seagreen')
plt.scatter(setosa['sepal length'],setosa['petal width'], label= 'Setosa', color='darkred')
plt.scatter(virginica['sepal length'],virginica['petal width'], label = 'Virginica', color='steelblue')
plt.legend()

plt.xlabel('sepal length')
plt.ylabel('petal width')
plt.show()
````

##### **Gráfico de Linhas**

````python
# Versicolor
plt.figure(figsize=(12,5))

plt.plot(versicolor['petal width'], label='petal width', marker='o')
plt.plot(versicolor['petal length'], label='petal length', marker='o')
plt.plot(versicolor['sepal width'], label='sepal width', marker='o')
plt.plot(versicolor['sepal length'], label='sepal length', marker='o')

plt.grid(linewidth=0.5, color = 'lightgray')
plt.ylim(0,8)

plt.legend(loc = 'upper center', ncols= 4)
plt.title('Características Iris-Versicolor')

plt.show()
````

````python
# Virginica
plt.figure(figsize=(12,5))

plt.plot(virginica['petal width'], label='petal width', marker='o', color='blue')
plt.plot(virginica['petal length'], label='petal length', marker='o',color='red')
plt.plot(virginica['sepal width'], label='sepal width', marker='o', color='orange')
plt.plot(virginica['sepal length'], label='sepal length', marker='o', color='gr')


plt.grid(linewidth=0.5, color = 'lightgray')
plt.ylim(0,10)

plt.legend(loc = 'upper center', ncols= 4)
plt.title('Características Iris-Virginica')

plt.show()
````

````python
# Setosa
plt.figure(figsize=(12,5))

plt.plot(setosa['petal width'], label='petal width', marker='o')
plt.plot(setosa['petal length'], label='petal length', marker='o')
plt.plot(setosa['sepal width'], label='sepal width', marker='o')
plt.plot(setosa['sepal length'], label='sepal length', marker='o')

plt.grid(linewidth=0.5, color = 'lightgray')
plt.ylim(0,7.5)

plt.legend(loc = 'upper center', ncols= 4)
plt.title('Características Iris-Setosa')

plt.show()
````
